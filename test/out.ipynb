```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label="sin(x)", color="#FF5A5F", linewidth=2)
plt.plot(x, np.cos(x), label="cos(x)", color="#00A699", linewidth=2)
plt.title("Sine and Cosine Waves")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
